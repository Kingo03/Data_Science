```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns
```

```python
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

```python
le = LabelEncoder()
df['letter'] = le.fit_transform(df['letter'])
```

```python
X_scaled = df.drop('letter', axis=1)
y = df['letter']
```

```python
from tensorflow.keras.utils import to_categorical
y_encoded = to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden layer 1
    Dense(32, activation='relu'),                                   # Hidden layer 2
    Dense(26, activation='softmax')                                 # Output layer (26 letters)
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_split=0.2, verbose=1)
# Test Accuracy: 93.25 for epochs = 30 and batch_size = 32 , 20% validation verbose = 1
# for epochs = 50 there is not change in the accuracy like 93.73%
#for epochs = 100 --Test Accuracy: 94.70%       /// best of all///
#lets now try for batch_size manipulation
# 1. batch size = 64  Test Accuracy: 94.55 epochs = 30
# 2. batch_size = 128 Test Accuracy: 94.25 decline from 64 size epochs = 50
# 3. batch_size = 64 and epochs = 100  - Test Accuracy: 94.48
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_acc:.4f}")
print(f'Test Loss: {test_loss:.4f}')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Predicted and actual classes
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)

# Display first 10 predictions
print("Predicted letters:", le.inverse_transform(y_pred_classes[:10]))
print("Actual letters:", le.inverse_transform(y_true[:10]))

# Calculate metrics
precision = precision_score(y_true, y_pred_classes, average='macro')
recall = recall_score(y_true, y_pred_classes, average='macro')
f1 = f1_score(y_true, y_pred_classes, average='macro')

print("\n--- Default Model Evaluation ---")
print("Test Accuracy : 0.9417")  # your given accuracy
print(f"Precision     : {precision:.4f}")
print(f"Recall        : {recall:.4f}")
print(f"F1-Score      : {f1:.4f}")

# Optional: Detailed classification report (for report submission)
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred_classes))
```

```python
import keras_tuner as kt
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# This function will tell KerasTuner which parameters to explore.
def build_model(hp):
    model = Sequential()
    
    # Tune number of hidden layers (1 to 3)
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(Dense(
            units=hp.Int(f'units_{i}', min_value=32, max_value=128, step=32),
            activation=hp.Choice('activation', values=['relu', 'tanh'])
        ))
    
    # Output layer (26 classes)
    model.add(Dense(26, activation='softmax'))
    
    # Tune learning rate
    lr = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,             # Try 5 combinations
    executions_per_trial=1,   # Each configuration is trained once
    directory='tuner_results',
    project_name='alphabet_ann_tuning'
)
```

```python
tuner.search(X_train, y_train, epochs=20, validation_split=0.2, verbose=1)
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"""
Best Hyperparameters:
1.Hidden Layers: {best_hps.get('num_layers')}
2.Neurons per Layer: {[best_hps.get(f'units_{i}') for i in range(best_hps.get('num_layers'))]}
3.Activation Function: {best_hps.get('activation')}
4.Learning Rate: {best_hps.get('learning_rate')}
""")
```

```python
best_model = tuner.hypermodel.build(best_hps)

history = best_model.fit(X_train, y_train, epochs=30, validation_split=0.2, verbose=1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

y_pred = np.argmax(best_model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)

acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average='macro')
rec = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')

print(f"Test Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"F1-Score: {f1:.4f}")
```
