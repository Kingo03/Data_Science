```python
# lets calculate measures of centrality for the numerical columns
# First let's calculate mean:
print("Mean for Volume:", df['Volume'].mean())
print("Mean for Avg Price:", df['Avg Price'].mean())
print("Mean for Total Sales Value:", df['Total Sales Value'].mean())
print("Mean for Discount Rate (%):", df['Discount Rate (%)'].mean())
print("Mean for Discount Amount:", df['Discount Amount'].mean())
print("Mean for Net Sales Value:", df['Net Sales Value'].mean())

```

```python
# Second Let's calculate median:
print("Median for Volume:", df['Volume'].median())
print("Median for Avg Price:", df['Avg Price'].median())
print("Median for Total Sales Value:", df['Total Sales Value'].median())
print("Median for Discount Rate (%):", df['Discount Rate (%)'].median())
print("Median for Discount Amount:", df['Discount Amount'].median())
print("Median for Net Sales Value:", df['Net Sales Value'].median())
```

```python
# mean > median therefore dataset is Rightly Skewed for all num_columns.but volume is slightly rightly skewed compare to others.
```

```python
# Third let's calcualte Mode:
print("Mode for Volume:", df['Volume'].mode())
print("Mode for Avg Price:", df['Avg Price'].mode())
print("Mode for Total Sales Value:", df['Total Sales Value'].mode())
print("Mode for Discount Rate (%):", df['Discount Rate (%)'].mode())
print("Mode for Discount Amount:", df['Discount Amount'].mode())
print("Mode for Net Sales Value:", df['Net Sales Value'].mode())
# Only Volume as single mode (3) all other have multiple mode so the data is descrete for all other 4 nums_columns 
# Discount Amount and Net Sales Value don't have frequent value in the dataset. Therefore, all 450 rows have descrete value for this columns. 
```

```python
# Fourth let's calculate Standard Deviation
print("SD for Volume:", df['Volume'].std())
print("SD for Avg Price:", df['Avg Price'].std())
print("SD for Total Sales Value:", df['Total Sales Value'].std())
print("SD for Discount Rate (%):", df['Discount Rate (%)'].std())
print("SD for Discount Amount:", df['Discount Amount'].std())
print("SD for Net Sales Value:", df['Net Sales Value'].std())
# Standard Deviation tells how spread out data is.
# tells how consistent a customer is with his buys
# Detecting outliers
```

```python
# Data Visualization
# Objective: 
# To visualize the distribution and relationship of numerical and 
# categorical variables in the dataset.
import seaborn as sns
import matplotlib.pyplot as plt

# Create subplots: 4 rows (one for each column), 2 columns (hist + boxplot)
fig, axes = plt.subplots(nrows=len(num_columns), ncols=2, figsize=(14, 18))
fig.suptitle("Data Visualization: Distribution and Relationships", fontsize=16)

# Loop for histograms and boxplots
for i, col in enumerate(num_columns): #enumerate() function adds a counter to each item in a list or other iterable. Here it allows indexing of the num_columns
    # Histogram
    sns.histplot(df[col], kde=True, ax=axes[i][0], color='skyblue')
    axes[i][0].set_title(f'Histogram of {col}')
    
    # Boxplot
    sns.boxplot(x=df[col], ax=axes[i][1], color='orange')
    axes[i][1].set_title(f'Boxplot of {col}')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define intended categorical columns
intended_cat_cols = ['Day', 'SKU', 'BU', 'Brand', 'Model'] 

# Filter only existing columns from the dataset
cat_columns = [col for col in intended_cat_cols if col in df.columns]

# Create subplots
n = len(cat_columns)
rows = (n + 1) // 2
fig, axes = plt.subplots(nrows=rows, ncols=2, figsize=(16, 5 * rows))
axes = axes.flatten()

# Plot bar charts for each categorical column
for i, col in enumerate(cat_columns):
    sns.countplot(data=df, x=col, hue=col, ax=axes[i], palette='pastel', legend=False)
    axes[i].set_title(f'Category Count: {col}')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].grid(axis='y', linestyle='--', alpha=0.6)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# #Standardization of Numerical Variables
# # 	Objective: To scale numerical variables for uniformity, improving the datasets suitability for analytical models.
# 	Explain the concept of standardization (z-score normalization).
# 	Standardize the numerical columns using the formula: z=x-mu/sigma
	
# 	Show before and after comparisons of the data distributions.
from sklearn.preprocessing import StandardScaler
import pandas as pd

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply z-score formula manually
df_standardized = df.copy()
df_standardized[num_cols] = (df[num_cols] - df[num_cols].mean()) / df[num_cols].std()

print(df_standardized.head(2))

import matplotlib.pyplot as plt

# Boxplots after
df_standardized[num_cols].plot(kind='box', subplots=True, layout=(2,3), figsize=(12,8), title="After Standardization")
plt.show()
```

```python
# Before Standardization  each column has different ranges.

# After Standardization  all columns are rescaled and centered around 0.
```

```python
# # Conversion of Categorical Data into Dummy Variables
# 	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
# 	Steps:
# 	Discuss the need for converting categorical data into dummy variables (one-hot encoding).
# 	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
# 	Display a portion of the transformed dataset.

# Select categorical columns
intended_cat_cols = ['Day', 'SKU','City', 'BU', 'Brand', 'Model'] 

print("Categorical columns:", intended_cat_cols)

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=intended_cat_cols, drop_first=True)

# Display transformed dataset
print(df_encoded.head(1))

df_encoded
```
