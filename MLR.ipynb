```python
# Model 4

df.info()
```

```python
x = df[['Fuel Type', 'Automatic', 'Doors', 'Gears','Age_08_04','KM', 'HP', 'cc', 'Cylinders', 'Weight']]
y = df['Price']

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
x_train = x_train.values
x_train 
```

```python
y_train = y_train.values.reshape(-1, 1)
y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
x_test = x_test.values
x_test
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predict on test data
y_pred = model.predict(x_test)

# Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R:", r2)
print(mean_absolute_percentage_error(y_test,y_pred)*100)
```

```python
from sklearn.linear_model import Ridge, Lasso
ridge = Ridge(alpha=1.0)   # alpha = regularization strength
ridge.fit(x_train, y_train)

y_pred_ridge = ridge.predict(x_test)

# Evaluation
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression:")
print("RMSE:", rmse_ridge)
print("MAE:", mae_ridge)
print("R:", r2_ridge)
print(mean_absolute_percentage_error(y_test,y_pred_ridge)*100)


```

```python
lasso = Lasso(alpha=0.01)   # smaller alpha keeps more features
lasso.fit(x_train, y_train)

y_pred_lasso = lasso.predict(x_test)

# Evaluation
rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("\nLasso Regression:")
print("RMSE:", rmse_lasso)
print("MAE:", mae_lasso)
print("R:", r2_lasso)
print(mean_absolute_percentage_error(y_test,y_pred_lasso)*100)
```
