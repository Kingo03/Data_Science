```python
# Required Libraries
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# X = df.drop("Target", axis=1)  # remove target variable if we have it our dataset otherwise no need .here we don't need it
X = df
```

```python
# 1. Standardize (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.mean() #  -ve means  0
```

```python
X_scaled.std()
```

```python
# 2. Apply PCA (let’s keep 2 components for visualization)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# 3. Create a new DataFrame
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])

print(pca_df.head())
```

```python
pca_df.count
```

```python
# 4. Explained variance (how much info each PC keeps)
explained_var =pca.explained_variance_ratio_
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

plt.scatter(pca_df["PC1"], pca_df["PC2"])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Result")
plt.show()

# Scree plot
plt.plot(np.cumsum(explained_var))
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance by PCA Components")
plt.grid(True)
plt.show()
```

```python
# Import libraries
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score


# Assuming df contains your original features
Y = df# convert to numpy array

# Step 1: Apply K-Means
k = 3  # choose number of clusters (can use elbow method to find optimal k)
kmeans_orig = KMeans(n_clusters=k, random_state=42)
labels_orig = kmeans_orig.fit_predict(Y)

# Step 2: Visualization (2D scatter using first two features)
plt.figure(figsize=(6,4))
sns.scatterplot(x=Y[:,0], y=Y[:,1], hue=labels_orig, palette='Set2', s=80)
plt.title('K-Means Clustering on Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Step 3: Evaluate clustering
sil_score_orig = silhouette_score(Y, labels_orig)
db_score_orig = davies_bouldin_score(Y, labels_orig)

print(f"Silhouette Score (Original Data): {sil_score_orig:.3f}")
print(f"Davies-Bouldin Index (Original Data): {db_score_orig:.3f}")
```

```python
from sklearn.decomposition import PCA

# Step 1: PCA transformation
pca = PCA(n_components=2)  # reduce to 2 components for easy visualization
X_pca = pca.fit_transform(X)

# Apply K-Means on PCA-transformed data
kmeans_pca = KMeans(n_clusters=k, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca)

# Step 2: Visualization
plt.figure(figsize=(6,4))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=labels_pca, palette='Set2', s=100)
plt.title('K-Means Clustering on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Step 3: Evaluate clustering
sil_score_pca = silhouette_score(X_pca, labels_pca)
db_score_pca = davies_bouldin_score(X_pca, labels_pca)

print(f"Silhouette Score (PCA Data): {sil_score_pca:.3f}")
print(f"Davies-Bouldin Index (PCA Data): {db_score_pca:.3f}")
```
